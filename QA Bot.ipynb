```python
from langchain.document_loaders import OnlinePDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

file = OnlinePDFLoader("https://d3bxy9euw4e147.cloudfront.net/oscms-qa/media/documents/IntroductionToBusiness-OP.pdf")
doc = file.load()

file_split = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
new_files = file_split.split_documents(doc)
```

```python
from langchain.vectorstores import Chroma,Pinecone
from langchain.embeddings.openai import OpenAIEmbeddings
import pinecone

embedding = OpenAIEmbeddings(openai_api_key="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")
```

```python
pinecone.init(
    api_key="a7584aad-cb06-43ce-8e2b-a4b7e4871140",
    environment="gcp-starter"
)
name = "qa-bot"
```

```python
pine_search = Pinecone.from_texts([sentence.page_content for sentence in new_files], embedding, index_name=name)
```

```python
from langchain.llms import OpenAI
from langchain.chains.question_answering import load_qa_chain

llm = OpenAI(temperature=0,openai_api_key="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")
chain = load_qa_chain(llm,chain_type="stuff")
```

```python
query = "What do you mean by business?"
answer = pine_search.similarity_search(query)
```

```python
chain.run(input_documents=answer,question=query)
```
